Perceptron Kernel
=================

Imports and plot inline.

```python
%matplotlib inline

import happyml as ml
```

Loading dataset from CSV.

```python
dataset = ml.datasets.load("circle.csv")

ml.plot.set_theme("minus_plus")
dataset.plot()
```

Create and train a perceptron with a **gaussian kernel** with $\sigma=1$ by default.

```python
perceptron = ml.models.PerceptronKernel(dataset.X, dataset.Y)

# Train perceptron kernel using Perceptron Learning Algorithm.
def train_and_plot(max_iterations=10):
    iterations = perceptron.pla(max_iterations)

    print "Training finished after", iterations, "iterations."
    print "Error in training set:", perceptron.error(dataset.X, dataset.Y) * 100, "%"
    
    dataset.plot()
    perceptron.plot()

train_and_plot()
```

Every point is predicted as positive. Let's change the value of $\sigma$ and test another kernels.

```python
kernel = ml.kernels.create_gaussian(sigma=0.5)
perceptron.kernel = kernel

train_and_plot()
```

```python
kernel = ml.kernels.create_gaussian(sigma=0.2)
perceptron.kernel = kernel

train_and_plot()
```

Polynomial kernel: $K(x_1, x_2) = (x_1 \cdot x_2 + c)^d$

```python
kernel = ml.kernels.create_polynomial(c=0.5, d=2)
perceptron.kernel = kernel

train_and_plot(1000)
```
