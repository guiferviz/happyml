Linear Regression
=================

Imports and plot inline.

```python
%matplotlib inline

import numpy as np

import happyml as ml
```

Loading a very simple regression dataset.

```python
dataset = ml.datasets.load("parabola.csv")

dataset.plot()
```

Linear Regression
-----------------

The hypothesis has the following form:

$h(x) = x \cdot w + b$

```python
model = ml.models.LinearRegression()

model.fit(dataset.X, dataset.Y)

# Plotting.
dataset.plot()
model.plot(title="$w=%.3f, b=%.3f$" % (model.w, model.b), title_size=20)
```

Transformations
---------------

In this example, we add a new column to our dataset with the square of the $x$ coordinate. The hypothesis has the following form:

$h(x) = x^2 \cdot w_2 + x \cdot w_1 + b$

```python
model2 = ml.models.LinearRegression(d=2)
model2.plot_type = "line"

def transform(X):
    return np.c_[X, np.square(X[:, 0])]

X = transform(dataset.X)
print "MSE Error (before fit):", model2.mse(X, dataset.Y)
model2.fit(X, dataset.Y)

dataset.plot()
model2.plot(title="$\\mathbf{w}=[%.3f, %.3f], b=%.3f$" % (model2.w[0], model2.w[1], model2.b),
            title_size=20,
            transform=transform)

print "MSE Error (after fit):", model2.mse(X, dataset.Y)
```
