Perceptron
==========

Imports and plot inline.

```python
%matplotlib inline

import happyml as ml
```

Loading dataset from CSV.

```python
dataset = ml.datasets.load("linearly_separable.csv")
```

Plotting classification dataset.

```python
# Set theme.
ml.plot.set_theme("minus_plus")

dataset.plot(limits=[-1, 1, -1, 1],
             grid=True,
             xlabel="$x_1$",
             ylabel="$x_2$",
             label_size=20,
             title="Linearly separable dataset",
             title_size=15)
```

Create and train perceptron.

```python
perceptron = ml.models.Perceptron()

# Train perceptron using Perceptron Learning Algorithm.
iterations = perceptron.pla(dataset.X, dataset.Y)

print "Training finished after", iterations, "iterations."
print "Error in training set:", perceptron.error(dataset.X, dataset.Y) * 100, "%"
```

Plotting dataset and decision boundary.

```python
# Show weights and bias term on the title.
title = r'$\mathbf{w} = [%0.2f, %0.2f]$  $b=%0.2f$' % (
    perceptron.w[0],
    perceptron.w[1],
    perceptron.b)

dataset.plot()
perceptron.plot(xlabel="$x_1$",
                ylabel="$x_2$",
                label_size=20,
                title=title,
                title_size=18)
```

Plot feasible region of $w_1$ and $w_2$. Any combination of weights that falls in the green zone is a pair of weights that classifies well all points in the dataset.

In order to obtain a simple 2D display, the bias term has been set to 0.

```python
class FeasibleWeights(ml.models.Model):
    plot_type = "binary_ones"
    def h(self, w):
        p = ml.models.Perceptron()
        p.w = w
        p.b = 0
        return -1 if p.accuracy(dataset.X, dataset.Y) < 1 else 1

fw = FeasibleWeights()
fw.plot(samples=200,  # Add samples to avoid sharpened lines (by default 50 samples per axis are used).
        title="Feasible region",
        xlabel="$w_1$",
        ylabel="$w_2$",
        label_size=20)
```
